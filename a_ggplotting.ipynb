# A step by step building of a GGplot

In this lab, we show how to build, step by step, a ggplot.
There exist several way for doing the same thing with ggplot (This may be an advantage?).
For example, using the car dataset, I can plot a scatterplot using alternatively these codes.


```r
# Libraries
library(ggplot2) #load the ggplot library
library(gridExtra) # Load two libraries for arranging plots
library(grid)
library(repr)


# Data
data(mtcars)
head(mtcars)
```

```r
options(repr.plot.width=6, repr.plot.height=4)
```

```r
# Six different ways for ployying the same things
p1 <- qplot(x = mpg,y=hp, data=mtcars,geom = "point")+ggtitle("1") 
p2 <- ggplot(data=mtcars, aes(x=mpg,y=hp))+geom_point()+ggtitle("2")
p3 <- ggplot(data=mtcars)+geom_point(aes(x=mpg,y=hp))+ggtitle("3")
p4 <- ggplot()+geom_point(data=mtcars,aes(x=mpg,y=hp))+ggtitle("4")
p5 <- ggplot(data=mtcars, aes(x=mpg,y=hp))+
      layer(geom="point",stat = "identity",position="identity")+ggtitle("5")
p6 <- ggplot(mtcars, aes(mpg,hp))+
      layer(geom="point",stat = "identity",position="identity")+ggtitle("6")
grid.arrange(p1,p2,p3,p4,p5,p6,nrow=2)
```

## Main elements for building a ggplot

For a **ggplot** we need five components of a layer:

-  The data, which must be an **R data frame**, and can be changed after the plot is created.(**NOTE: data must be a dataframe or a tibble**)

-  A set of **aesthetic mappings**,  which describe how variables in the data are mapped to aesthetic properties of the layer. This section includes a description of how layer settings override the plot defaults, the difference between setting and mapping, and the important group aesthetic.
- The **geom** which describes the geometric used to draw the layer. 
   + The geom defines the set of available **aesthetic properties**.
   

- The **stat**,  which takes the raw data and transforms it in some useful way. The stat returns a data frame with new variables that can also be mapped to aesthetics with a special syntax.
- The **position** adjustment which adjusts elements to avoid overplotting.

`layer(geom="point",stat = "identity",position="identity")`

is the **verbose** version, but is useful to undertand it because, a layer has a **geom** part, but also a **stat** and a **position**.

Usually, it is better to use ***shortcut functions*** having the same basic form, beginning with **geom_** or **stat_**:

> `geom_XXX(mapping, data, ..., geom, position)`  
> `stat_XXX(mapping, data, ..., stat, position)`

> `geom_XXX(mapping, data, ..., geom, position)`  
> `stat_XXX(mapping, data, ..., stat, position)`  

#### Their common parameters define the components of the layer:  

- **mapping** (optional): A set of aesthetic mappings, specified using the aes() function and combined with the plot defaults.
- **data** (optional): A dataset which overrides the default plot dataset. It is most commonly omitted, in which case the layer will use the default plot
data. 
- **...**: Parameters for the ***geom*** or ***stat***, such as bin width in the histogram or bandwidth for a loess smoother. You can also use aesthetic properties as parameters. When you do this you set the property to a fixed value, not map it to a variable in the dataset.
-  **geom** or **stat** (optional): You can override the default stat for a geom, or the default geom for a stat. This is a text string containing the name of the geom to use. Using the default will give you a standard plot; overriding the default allows you to achieve something more exotic.
-  **position** (optional): Choose a method for adjusting overlapping objects.

### Help of geom_point: Aestetichs
***geom_point()*** understands the following aesthetics (required aesthetics are in bold):
from https://ggplot2.tidyverse.org/reference/geom_point.html
- **x** x-axis
- **y** y-axis
- alpha *transparency (a value between 0 to 1)*
- colour *colour of border of shapes*
- fill *Some shapes have borders (see shape=21)*
- group *overrides the standard way of grouping points (a minor aes) *
- shape *a number corresponding to different shapes* see http://sape.inf.usi.ch/quick-reference/ggplot2/shape
- size *a number >0 for the size of points*
- stroke *fo some shapes you can modify the borders or the stroke of its traits* see https://ggplot2.tidyverse.org/articles/ggplot2-specs.html

```r
options(repr.plot.width=10, repr.plot.height=4)
```

```r
p1 <- ggplot(mtcars, aes(mpg, hp, colour=as.factor(cyl))) + geom_point(aes(size=cyl))+ggtitle("geom_point(aes(size=cyl))")
p2 <- ggplot(mtcars, aes(mpg, hp, colour=as.factor(cyl))) + geom_point(size=3)+ggtitle("geom_point(size=3)")
grid.arrange(p1,p2,ncol=2)
```
